```python
def random(n=1,seed=None):
    '''Generates Random number.
    n = int | default = 1 (when 1 returns single random number, else returns array of n random numbers)
    seed = int | default = None'''
    
    # Declaring two variable one random and second double mersenne prime number
    a=25879
    b=(2**31)-1
    
    # conditional statement for seed
    if seed is None:
        # choosing id of empty list as seed as it will be different each time
        seed=id([])
    else:
        # multiplying seed with a high number
        seed=seed*123456789
    # Generating a number with math  
    x=(seed*a+1)%b
    result=[0]*n
    result[0]=x/b
    for i in range(1,n):
        x=(x*a+1)%b
        result[i]=x/b
    # Conditional statement for output
    if n==1:
        return result[0]
    else:
        return result
```

The above figure is plotted using 10000 random points on x and y axis using random function created above. where r is the distance of a point from origin(0,0).If r is less than 1 then the point lies in the circle else in the square.

<center>Area of circle  $C = \pi r^2$ <br>
Area of square $S = side^2$ </center><br>
<i> in the figure above their is $^1/_4$ <sup>th</sup> of each square and circle.<br>
and circle's radius (r) is equal to square's side (s) </i><br>
$So,$ <br>
<center>
$S=r^2$ ........(1) <br>
Area in orange $O = \frac{\pi r^2}{4}$ ........(2) <br>
<br>
$\pi = \frac{4 * O}{r^2}$ ........from(2) <br>
$\pi = \frac{4*O}{S}$ ........substituting (1)<br>
</center>
Area in orange $O$ is equal to total number of points with r less than 1 <br>
Area of square $S$ or Total area is equal to total number of points<br>

$So,$<br>
$$Value of  \pi = \frac{4 * number of points with r<1 }{Total number of points}$$
<br><br>
Implementing this with a function,

```python
def get_pi():
    n=1000000
    x=random(n)
    y=random(n)
    points_in_circle = 0
    for i,j in list(zip(x,y)):
        r = (i**2 + j**2)**(1/2)
        if r<1:
            points_in_circle+=1
    value = 4*points_in_circle/n
    print('{:.2f}'.format(value))
```

```python
get_pi()
```

<i><sub>More the n in above function more the accuracy.</sub></i>
